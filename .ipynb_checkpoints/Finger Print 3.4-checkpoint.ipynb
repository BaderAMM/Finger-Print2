```python
from bs4 import BeautifulSoup
import pandas as pd
import time
import requests
import datetime
```

```python
file = 'Month 01_3.html'
# soup = BeautifulSoup(file, 'HTML')


with open(file,encoding="utf-8") as fp:
    soup = BeautifulSoup(fp, 'lxml')
```

```python
############### Join ############
df = pd.DataFrame()
dic = {'Day':'cs6E0B0C5C',
       'Date':'cs895374D5',
      'inter':9,
       'out':'cs63677036',
       'late':'csF1939D0B',
       'short':'cs39D507AE',
       'Add':'cs8D14490E',
       'Jazaa':4,
      'work Hour':3,
       'Note':'cs736D34A6',
       'shift':'cs41F7A28F'
       }
dic['shift'] # value

keys = list(dic.keys())
for key in keys:
    lis1 = []
    #================
    if key in  ['inter','work Hour','Jazaa']:
        trs = soup.find_all("tr", style="vertical-align:top;")

        for tr in trs[18:]:
            lenth = len(tr.find_all('td'))
            if lenth == 13:
                td_text = tr.find_all('td')[dic[key]]
                td_text_skip = tr.find_all('td')[9]
                
                if td_text_skip['class'] == ['cs10FC340C'] : # cs10FC340C for weekend  A == 1 or 
                    continue
                lis1.append(td_text.get_text())
        
    else:
        dates = soup.find_all("td", class_=dic[key])#.find_all('nobr')[0].get_text()
        
        for date in dates:
            date1 = date.get_text()
            lis1.append(date1)

            
    s = pd.Series(lis1)
    df[key] = s
# df
```

```python
# 'csC8E988B9'         Total Delay & Shortage Hours

Tdelay = soup.find_all("td", class_='csC8E988B9')
delayTimeText = Tdelay[0].get_text()
print(delayTimeText)
delayTime = datetime.datetime.strptime(delayTimeText, "%H:%M")
delayTime
```

```python
# df2 = pd.DataFrame()
cols = [df['late'],df['short']]
a = 0
for col in cols:
   
    lisTime = []
    for val in col:
        

        try:
            x = datetime.datetime.strptime(val, "%H:%M")
            lisTime.append(x)
        except:
            val = '00:00'
            x = datetime.datetime.strptime(val, "%H:%M")
            lisTime.append(x)

#     df2['test'] = lisTime
    
#     asd = df2['test'].dt.time.sort_values(ascending=False).reset_index() #,ignore_index=True
#     print(asd['test'])
#     df2['tezt1'] = asd['test']
    
    nameCol = ['TopDelay','TopShort']

    df[nameCol[a]] = lisTime #.dt.time.sort_values(ascending=False)
    df[nameCol[a]] = df[nameCol[a]].dt.time

    a = a +1

df
```

```python
dfDelay = (df[['Day','Date','TopDelay']]).rename(columns={'TopDelay':'LATE'})
dfShort = (df[['Day','Date','TopShort']]).rename(columns={'TopShort':'LATE'})
# dfDelay.rename{'TopDelay'}
# dfTotal2 = 
dfShort['Whene'] = 'Short'
dfShort

dfDelay['Whene'] = 'Delay'
dfDelay

dfTotal2 = pd.concat([dfDelay,dfShort])
dfTotal2.sort_values(by=['LATE'],ascending=False, inplace=True)
dfTotal2.reset_index(inplace=True)
df_Top_Late = dfTotal2[:4]
df_Top_Late
```

```python
time1 = dfTotal2.loc[0,'LATE']#.total_seconds()
t_permision = 0
for time1 in dfTotal2[:3].loc[:,'LATE']:
    print(time1)
    
    timeobj = time1#time(12, 45)
    t = datetime.datetime.combine(datetime.date.min, timeobj) - datetime.datetime.min
#     print(datetime.date.min)
#     print(datetime.datetime.min)
#     print(datetime.datetime.combine(datetime.date.min, timeobj))
    isinstance(t, datetime.timedelta)
#     print(isinstance(t, datetime.timedelta))
    # True
    print('total_seconds',t.total_seconds())
    t_permision += t.total_seconds()
# 45900.0
t_permision


seconds_input = t_permision
conversion = datetime.timedelta(seconds=seconds_input)
print(conversion)
permision_hour = conversion
print('=============')
print(permision_hour)
print(str(permision_hour))

#output 12:01:30
# per = t_permision.datetime.time()
# per
```

```python
delayTime.time()
t_Late = datetime.datetime.combine(datetime.date.min, delayTime.time()) - datetime.datetime.min
t_Late = t_Late.total_seconds()
print(t_Late)
print(t_permision)
Work_hour = 8 * 60 * 60
print(Work_hour)

V_day_Second = (t_Late - t_permision)//Work_hour
V_day = V_day_Second/60/60
V_day
# 7.5//8
```

```python
dic_Report = {
    'total LATE':delayTime.time(),
    'total permision':str(permision_hour),
              'Vication day':V_day
             }
dic_Report
df_Report = pd.DataFrame(dic_Report, index = ['Time']) #dic_Report.keys()
df_Report.T

# df_Report = pd.DataFrame.from_dict(dic_Report, orient = 'index') #dic_Report.keys()
# df_Report
```
