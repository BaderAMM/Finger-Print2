```python
df = pd.DataFrame()
dic = {'Day':'cs6E0B0C5C',
       'Date':'cs895374D5',
       'inter':'csCCDF7E4B',
       'out':'cs63677036',
       'late':'csF1939D0B',
       'short':'cs39D507AE',
       'Add':'cs8D14490E',
       'Jazaa':'csCCDF7E4B',
       'work Hour':'csCCDF7E4B',
       'Note':'cs736D34A6',
       'shift':'cs41F7A28F'
       }
dic['shift']

for item in range(len(dic)):
    key = list(dic.keys())[item]
    print(key)


    dic[key]

    dates = soup.find_all("td", class_=dic[key])#.find_all('nobr')[0].get_text()
    print(len(dates))
#     if len(dates) != 20:
#         continue
    lis1 = []

    for date in dates:
        date1 = date.find_all('nobr')#[0].get_text()
    #     print(date1)
        lis1.append(date1)

    s = pd.Series(lis1)
    df[key] = s
df
```
